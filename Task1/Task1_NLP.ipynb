Task 2: Preprocessing

```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo -> soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters (keep words)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    return text

    # Remove very short tokens (<=2) except 'no', 'not'
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    cleaned_sentence = ' '.join(filtered_tokens)


```

```python
sentence = "I have 2 dogs"
processed = preprocess_text(sentence)

print("Original:", sentence)
print("Processed:", processed)
```

Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("\n--- Stress Testing Output ---\n")

for i, sentence in enumerate(sentences, 1):
    cleaned_text = preprocess_text(sentence)

    tokens = cleaned_text.split()

    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    cleaned_sentence = ' '.join(filtered_tokens)

    print(f"Sentence {i}:")
    print("Original Text:", sentence)
    print("Cleaned Tokens:", filtered_tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 40)
```

Task 4: Token Analytics

```python
print("\n--- Token Analytics ---\n")

for i, sentence in enumerate(sentences, 1):
    cleaned_text = preprocess_text(sentence)

    # Tokenization
    tokens = cleaned_text.split()

    # Remove short tokens (<=2) except 'no', 'not'
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    # Analytics
    total_tokens = len(filtered_tokens)
    unique_tokens = len(set(filtered_tokens))

    if total_tokens > 0:
        avg_token_length = sum(len(word) for word in filtered_tokens) / total_tokens
    else:
        avg_token_length = 0

    print(f"Sentence {i}:")
    print("Cleaned Tokens:", filtered_tokens)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_token_length, 2))
    print("-" * 40)
```

Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in sentences:
    cleaned_text = preprocess_text(sentence)


    tokens = cleaned_text.split()

    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    all_tokens.extend(filtered_tokens)

# Frequency Analysis
counter = Counter(all_tokens)

print("\n--- Frequency Analysis ---\n")

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
least_common = sorted(counter.items(), key=lambda x: x[1])[:5]
print(least_common)
```

Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        cleaned_text = preprocess_text(text)

        # Tokenization
        tokens = cleaned_text.split()

        filtered_tokens = [
            word for word in tokens
            if len(word) > 2 or word in ['no', 'not']
        ]

        all_tokens.extend(filtered_tokens)
        clean_sentences.append(' '.join(filtered_tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(sentences)
print(output)
```

Task 7: Error Handling Tests

```python
error_test_cases = [
    "",                 # Empty string
    "😂😂😂😂😂",        # Only emojis
    "1234567890"        # Only numbers
]

print("\n--- Error Handling Output ---\n")

for i, case in enumerate(error_test_cases, 1):
    result = preprocess_text(case)

    # Handle different return types safely
    if isinstance(result, tuple) and len(result) == 2:
        tokens, cleaned_sentence = result
    elif isinstance(result, str):
        tokens = result.split()
        cleaned_sentence = result
    else:
        tokens, cleaned_sentence = [], ""

    print(f"Test Case {i}:")
    print("Input:", repr(case))
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 40)
```
